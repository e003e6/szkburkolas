```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import networkx as nx

from shapely.ops import nearest_points, unary_union
from shapely.geometry import LineString, Polygon, MultiPolygon

# -----------------------
# PARAMTEREK
# -----------------------
PLACE = "Tkl, Hungary"      # Vizsglt telepls neve az OSM-ben
DIST_LIM = 200                # max tvolsg (mterben, projekciban) ameddig "rhzunk" a boundary-ra
MIN_SEG_LEN = 0.1             # minimlis szakaszhossz (m)

# -----------------------
# 1) THLZAT LETLTSE S PROJEKTLSA
# -----------------------
G = ox.graph_from_place(PLACE, network_type="drive")
Gp = ox.project_graph(G)
nodes_gdf, edges_gdf = ox.graph_to_gdfs(Gp, nodes=True, edges=True)

# Csak utcavgek (dead-end): fokszm 1 (undirected)

G_und = Gp.to_undirected()
deg = dict(G_und.degree())

dead_end_node_ids = [n for n, d in deg.items() if d == 1]
dead_ends_gdf = nodes_gdf.loc[nodes_gdf.index.intersection(dead_end_node_ids)].copy()

# -----------------------
# 2) lakvezet hatr letltse
# landuse = residential OSM
# -----------------------
tags = {"landuse": "residential"}
res_gdf = ox.features_from_place(PLACE, tags=tags)

# Csak poligonok
res_poly = res_gdf[res_gdf.geometry.type.isin(["Polygon", "MultiPolygon"])].copy()

if res_poly.empty:
    raise RuntimeError(
        "Nem talltam landuse=residential poligonokat ehhez a PLACE-hez az OSM-ben."
    )

# Projektls a grf CRS-be (mteres)
res_poly_proj = res_poly.to_crs(nodes_gdf.crs)

# Egyests + hatrvonal
from shapely.ops import unary_union
res_union = unary_union(res_poly_proj.geometry.values)
boundary_mls_proj = res_union.boundary

boundary_lines_proj = gpd.GeoSeries([boundary_mls_proj], crs=nodes_gdf.crs)

# -----------------------
# 3) CSOMPONTOK SSZEKTSE A HATRVONALLAL
# -----------------------
extensions_proj = []

for idx, row in dead_ends_gdf.iterrows():
    pt = row.geometry

    d = pt.distance(boundary_mls_proj)
    if pd.notna(d) and np.isfinite(d) and d <= DIST_LIM:
        _, near = nearest_points(pt, boundary_mls_proj)
        if near is not None and (not near.is_empty):
            seg = LineString([pt, near])
            if seg.length > MIN_SEG_LEN:
                extensions_proj.append(seg)

extensions_proj_gs = gpd.GeoSeries(extensions_proj, crs=nodes_gdf.crs)

# -----------------------
# 4) VISSZAVETTS WGS84-BE
# -----------------------
boundary_lines_wgs = boundary_lines_proj.to_crs(epsg=4326)
extensions_wgs = extensions_proj_gs.to_crs(epsg=4326) if len(extensions_proj_gs) else gpd.GeoSeries([], crs=nodes_gdf.crs).to_crs(4326)

# -----------------------
# 5) L-SZNEZS NV ALAPJN
# -----------------------
def has_name(val):
    if isinstance(val, list):
        return any(isinstance(i, str) and i.strip() for i in val)
    if pd.isna(val):
        return False
    return isinstance(val, str) and val.strip()

edge_colors = [
    "white" if has_name(data.get("name")) else "red"
    for _, _, _, data in G.edges(keys=True, data=True)
]

# -----------------------
# 6) RAJZOLS
# -----------------------
fig, ax = ox.plot_graph(
    G,
    node_size=0,
    edge_color=edge_colors,
    bgcolor="black",
    show=False,
    close=False
)

ax.set_aspect("auto")

if len(boundary_lines_wgs):
    boundary_lines_wgs.plot(ax=ax, color="lime", linewidth=2)

if len(extensions_wgs):
    extensions_wgs.plot(ax=ax, color="dodgerblue", linewidth=1.8)

# Zoom a lakterlet hatrra
if len(boundary_lines_wgs):
    minx, miny, maxx, maxy = boundary_lines_wgs.total_bounds
    pad = 0.002
    ax.set_xlim(minx - pad, maxx + pad)
    ax.set_ylim(miny - pad, maxy + pad)

plt.show()
```

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from shapely.geometry import Point, LineString

PLACE = "Tkl, Hungary"
MAX_EXT = 2000.0   # m
EPS = 0.25         # m (sajt node krnyke)

# 1) graph + projektlt gdf-ek
G = ox.graph_from_place(PLACE, network_type="drive")
Gp = ox.project_graph(G)
nodes, edges = ox.graph_to_gdfs(Gp, nodes=True, edges=True)

# 2) dead-end node-ok (undirected degree = 1)
deg = dict(Gp.to_undirected().degree())
dead = nodes.loc[[n for n, d in deg.items() if d == 1]]

# segd: metszs eredmnybl pontok
def pts(geom):
    if geom.is_empty: return []
    t = geom.geom_type
    if t == "Point": return [geom]
    if t == "MultiPoint": return list(geom.geoms)
    if t == "LineString": return [Point(geom.coords[0]), Point(geom.coords[-1])]
    if t in ("MultiLineString", "GeometryCollection"):
        out = []
        for g in geom.geoms: out += pts(g)
        return out
    return []

# segd: dead-endbl kifel mutat "ray"
def ray_from_deadend(node_id):
    pt = nodes.loc[node_id].geometry
    u, v, k, data = list(Gp.edges(node_id, keys=True, data=True))[0]
    geom = data.get("geometry")
    if geom is None:
        other = v if u == node_id else u
        geom = LineString([pt, nodes.loc[other].geometry])

    c = list(geom.coords)
    a0, a1 = Point(c[0]), Point(c[-1])
    # node melyik vgponton van?
    if pt.distance(a0) <= pt.distance(a1):
        a, b = c[0], c[1]
    else:
        a, b = c[-1], c[-2]

    dx, dy = (a[0]-b[0], a[1]-b[1])   # kifel
    n = (dx*dx + dy*dy) ** 0.5
    if n == 0: return None

    far = Point(pt.x + dx/n * MAX_EXT, pt.y + dy/n * MAX_EXT)
    return pt, LineString([pt, far])

# 3) trbeli index
sidx = edges.sindex

# 4) hosszabbtsok
ext = []
for node_id in dead.index:
    pt, ray = ray_from_deadend(node_id)
    if ray is None: 
        continue

    cand = edges.iloc[list(sidx.intersection(ray.bounds))]

    best_p, best_s = None, np.inf
    for (eu, ev, ek), row in cand.iterrows():
        if node_id in (eu, ev):   # ne a sajt lt tallja meg rgtn
            continue
        inter = ray.intersection(row.geometry)
        for p in pts(inter):
            s = ray.project(p)    # tvolsg a ray mentn
            if s <= EPS or s >= best_s:
                continue
            best_s, best_p = s, p

    if best_p is not None:
        ext.append(LineString([pt, best_p]))

ext = gpd.GeoSeries(ext, crs=nodes.crs).to_crs(4326)

# 5) plot
fig, ax = ox.plot_graph(G, node_size=0, edge_color="white", bgcolor="black",
                        show=False, close=False)
if len(ext):
    ext.plot(ax=ax, color="orange", linewidth=2.0)
plt.show()
```

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from shapely.geometry import Point, LineString
from shapely.ops import unary_union, nearest_points

# -----------------------
# PARAMTEREK
# -----------------------
PLACE = "Tkl, Hungary"

MAX_EXT = 200.0   # m: narancs hosszabbts max
EPS = 0.25         # m: sajt node krnyke (narancs metszsnl)
DIST_LIM = 100.0   # m: ha node ennyire van a lak-hatrtl, hzzuk r kkkel
MIN_SEG = 0.1      # m: minimlis szakaszhossz

# -----------------------
# 1) THLZAT
# -----------------------
G = ox.graph_from_place(PLACE, network_type="drive")
Gp = ox.project_graph(G)  # mteres CRS
nodes, edges = ox.graph_to_gdfs(Gp, nodes=True, edges=True)

# dead-end node-ok (undirected degree=1)
deg = dict(Gp.to_undirected().degree())
dead = nodes.loc[[n for n, d in deg.items() if d == 1]].copy()

# -----------------------
# segdek (narancshoz)
# -----------------------
def _pts(geom):
    if geom.is_empty: return []
    t = geom.geom_type
    if t == "Point": return [geom]
    if t == "MultiPoint": return list(geom.geoms)
    if t == "LineString": return [Point(geom.coords[0]), Point(geom.coords[-1])]
    if t in ("MultiLineString", "GeometryCollection"):
        out = []
        for g in geom.geoms: out += _pts(g)
        return out
    return []

def _ray_from_deadend(node_id):
    pt = nodes.loc[node_id].geometry
    u, v, k, data = list(Gp.edges(node_id, keys=True, data=True))[0]
    geom = data.get("geometry")

    if geom is None:
        other = v if u == node_id else u
        geom = LineString([pt, nodes.loc[other].geometry])

    c = list(geom.coords)
    a0, a1 = Point(c[0]), Point(c[-1])
    if pt.distance(a0) <= pt.distance(a1):
        a, b = c[0], c[1]
    else:
        a, b = c[-1], c[-2]

    dx, dy = (a[0] - b[0], a[1] - b[1])  # kifel
    n = (dx*dx + dy*dy) ** 0.5
    if n == 0: return None, None
    far = Point(pt.x + dx/n * MAX_EXT, pt.y + dy/n * MAX_EXT)
    return pt, LineString([pt, far])

# -----------------------
# 2) NARANCS: dead-end -> kvetkez utca
# -----------------------
sidx_edges = edges.sindex
orange = []

for node_id in dead.index:
    pt, ray = _ray_from_deadend(node_id)
    if ray is None:
        continue

    cand = edges.iloc[list(sidx_edges.intersection(ray.bounds))]

    best_p, best_s = None, np.inf
    for (eu, ev, ek), row in cand.iterrows():
        if node_id in (eu, ev):          # sajt l ne
            continue
        inter = ray.intersection(row.geometry)
        for p in _pts(inter):
            s = ray.project(p)           # tvolsg a ray mentn
            if s <= EPS or s >= best_s:
                continue
            best_s, best_p = s, p

    if best_p is not None:
        seg = LineString([pt, best_p])
        if seg.length > MIN_SEG:
            orange.append(seg)

orange = gpd.GeoSeries(orange, crs=nodes.crs)

# -----------------------
# 3) LAKOTT TERLET POLYGONOK + BOUNDARY (MINDEN relevns folt)
#   - nem csak 1-et tartunk meg
#   - minden olyan residential poligont megtartunk,
#     amely metszi az thlzatot (vonalhossz > 0 a poligonon bell)
# -----------------------
res = ox.features_from_place(PLACE, tags={"landuse": "residential"})
res = res[res.geometry.type.isin(["Polygon", "MultiPolygon"])].copy()
if res.empty:
    raise RuntimeError("Nincs landuse=residential poligon ehhez a PLACE-hez az OSM-ben.")

res_p = res.to_crs(nodes.crs)

# thlzat (projektlt) sszefslve 1 geometriba
roads_union = edges.geometry.union_all()

# residential komponensek sztszedse (MultiPolygon -> kln Polygonok)
polys = []
for g in res_p.geometry:
    if g.geom_type == "Polygon":
        polys.append(g)
    elif g.geom_type == "MultiPolygon":
        polys.extend(list(g.geoms))

if not polys:
    raise RuntimeError("A residential geometrikbl nem tudtam poligonokat kinyerni.")

# kivlaszts: csak azok, ahol az thlzat tnylegesen BELP (vonalhossz > 0)
keep_polys = []
for p in polys:
    inter = roads_union.intersection(p)
    score = getattr(inter, "length", 0.0)
    if np.isfinite(score) and score > 0:
        keep_polys.append(p)

if not keep_polys:
    # fallback: centroid-ot tartalmazk (ha valamirt length=0)
    c = roads_union.centroid
    keep_polys = [p for p in polys if p.contains(c)]
    if not keep_polys:
        raise RuntimeError("Nem talltam olyan residential poligont, amihez az thlzat tartozna.")

# egyestett (MultiPolygon) a kkhez + boundary a plothoz
res_area = unary_union(keep_polys)          # Polygon/MultiPolygon
boundary = res_area.boundary               # LineString/MultiLineString
boundary_gs = gpd.GeoSeries([boundary], crs=nodes.crs)

# opcionlis: kln poligonok is, ha ksbb kell
res_polys_gdf = gpd.GeoDataFrame(geometry=keep_polys, crs=nodes.crs)

print("Residential poligonok (ttal rintett):", len(keep_polys))

# -----------------------
# 4) KK: minden node (DIST_LIM-en bell) -> lakhatr (az SSZES megtartott folt hatra)
# -----------------------
blue = []
for node_id, row in nodes.iterrows():
    pt = row.geometry
    d = pt.distance(boundary)
    if np.isfinite(d) and d <= DIST_LIM:
        _, near = nearest_points(pt, boundary)
        if near is not None and (not near.is_empty):
            seg = LineString([pt, near])
            if seg.length > MIN_SEG:
                blue.append(seg)

blue = gpd.GeoSeries(blue, crs=nodes.crs)

# -----------------------
# 5) PLOT (WGS84)
# -----------------------
boundary_wgs = boundary_gs.to_crs(4326)
orange_wgs = orange.to_crs(4326) if len(orange) else gpd.GeoSeries([], crs=nodes.crs).to_crs(4326)
blue_wgs = blue.to_crs(4326) if len(blue) else gpd.GeoSeries([], crs=nodes.crs).to_crs(4326)

def has_name(val):
    if isinstance(val, list):
        return any(isinstance(i, str) and i.strip() for i in val)
    if pd.isna(val):
        return False
    return isinstance(val, str) and val.strip()

edge_colors = [
    "white" if has_name(data.get("name")) else "red"
    for _, _, _, data in G.edges(keys=True, data=True)
]

fig, ax = ox.plot_graph(
    G,
    node_size=0,
    edge_color=edge_colors,
    bgcolor="black",
    show=False,
    close=False
)

ax.set_aspect("auto")

# hatr (zld)  mr az sszes relevns folt boundary-ja
boundary_wgs.plot(ax=ax, color="lime", linewidth=2)

# narancs (dead-end -> kvetkez utca)
if len(orange_wgs):
    orange_wgs.plot(ax=ax, color="orange", linewidth=2.0)

# kk (node -> lakhatr)
if len(blue_wgs):
    blue_wgs.plot(ax=ax, color="dodgerblue", linewidth=1.8)

# zoom a lakhatrra
minx, miny, maxx, maxy = boundary_wgs.total_bounds
pad = 0.002
ax.set_xlim(minx - pad, maxx + pad)
ax.set_ylim(miny - pad, maxy + pad)

plt.show()
```

```python
from shapely.ops import unary_union, linemerge
from shapely.geometry import LineString, MultiLineString, GeometryCollection
import geopandas as gpd
import matplotlib.pyplot as plt

# Felttelezzk, hogy az elz cellbl ezek lteznek:
# edges (GeoDataFrame, projektlt CRS-ben)
# boundary_gs (GeoSeries, projektlt CRS-ben)  -- vagy boundary (shapely)
# orange (GeoSeries, projektlt CRS-ben)
# blue (GeoSeries, projektlt CRS-ben)

# --- kompatibilits: ha boundary_gs nincs, de boundary igen ---
if "boundary_gs" not in globals():
    if "boundary" in globals():
        boundary_gs = gpd.GeoSeries([boundary], crs=edges.crs)
    else:
        raise RuntimeError("Nincs boundary_gs s boundary sem a notebookban.")

# 1) Vonalak sszegyjtse
street_lines    = [g for g in edges.geometry if g is not None and not g.is_empty]
boundary_lines  = [g for g in boundary_gs.geometry if g is not None and not g.is_empty]

connector_lines = []
if "orange" in globals() and orange is not None and len(orange):
    connector_lines += [g for g in orange.geometry if g is not None and not g.is_empty]
if "blue" in globals() and blue is not None and len(blue):
    connector_lines += [g for g in blue.geometry if g is not None and not g.is_empty]

all_lines_proj = street_lines + boundary_lines + connector_lines
print("sszes vonal:", len(all_lines_proj))

if not all_lines_proj:
    raise RuntimeError("Nincs egyesthet vonal (all_lines_proj res).")

def extract_lines(geom):
    """Kiszed minden LineString-et / MultiLineString-et egy tetszleges geometribl."""
    if geom is None or geom.is_empty:
        return []
    if isinstance(geom, LineString):
        return [geom]
    if isinstance(geom, MultiLineString):
        return list(geom.geoms)
    if isinstance(geom, GeometryCollection):
        out = []
        for g in geom.geoms:
            out.extend(extract_lines(g))
        return out
    return []

# 2) Topolgiai egyests (ha sikerl)
try:
    union_geom  = unary_union(all_lines_proj)
    union_lines = extract_lines(union_geom)
    print("unary_union utn vonalak:", len(union_lines))

    if union_lines:
        merged_geom  = linemerge(union_lines)
        merged_lines = extract_lines(merged_geom)
        print("linemerge utn vonalak:", len(merged_lines))
    else:
        merged_lines = []
except Exception as e:
    print("Hiba a unary_union / linemerge sorn:", repr(e))
    union_lines  = []
    merged_lines = []

# 3) Vgs hlzat kivlasztsa
if merged_lines:
    final_lines = merged_lines
    print("Vgs hlzat: merged_lines")
elif union_lines:
    final_lines = union_lines
    print("Vgs hlzat: union_lines")
else:
    final_lines = all_lines_proj
    print("Vgs hlzat: all_lines_proj (nem sikerlt az egyests)")

network_gs_proj = gpd.GeoSeries(final_lines, crs=edges.crs)
print("geom_type:\n", network_gs_proj.geom_type.value_counts())
print("total_bounds:", network_gs_proj.total_bounds)

# 4) Plot  CSAK az egyestett / kivlasztott vonalhlzat
fig, ax = plt.subplots(figsize=(10, 10))

network_gs_proj.plot(ax=ax, linewidth=2.0, color="red")
ax.set_aspect("equal")
ax.set_facecolor("0.2")
ax.grid(True, color="0.4", linestyle="--", linewidth=0.5)

minx, miny, maxx, maxy = network_gs_proj.total_bounds
side = max(maxx - minx, maxy - miny)
pad  = side * 0.05
ax.set_xlim(minx - pad, maxx + pad)
ax.set_ylim(miny - pad, maxy + pad)

plt.tight_layout()
plt.show()
```

```python
from shapely.ops import unary_union, linemerge
from shapely.geometry import LineString, MultiLineString, GeometryCollection
import geopandas as gpd
import matplotlib.pyplot as plt

# Kell:
# edges (GeoDataFrame, projektlt CRS)
# orange (GeoSeries, projektlt CRS) - opcionlis
# blue (GeoSeries, projektlt CRS)   - opcionlis
#
# s az elz cellbl valamelyik:
#  - res_area (Polygon/MultiPolygon)  <-- EZ A LEGJOBB
#  - vagy res_polys_gdf (GeoDataFrame a keep_polys-szal)
#  - vagy boundary_gs + keep_polys (ha gy nevezted)

# --- boundary GeoSeries biztostsa ---
if "boundary_gs" not in globals():
    if "boundary" in globals():
        boundary_gs = gpd.GeoSeries([boundary], crs=edges.crs)
    else:
        raise RuntimeError("Nincs boundary_gs s boundary sem a notebookban.")

# --- CLIP POLY: az SSZES relevns lakott terlet unija ---
if "res_area" in globals() and res_area is not None:
    clip_poly = res_area
elif "res_polys_gdf" in globals() and res_polys_gdf is not None and len(res_polys_gdf):
    clip_poly = unary_union(res_polys_gdf.geometry)
elif "keep_polys" in globals() and keep_polys is not None and len(keep_polys):
    clip_poly = unary_union(keep_polys)
else:
    raise RuntimeError(
        "Nem tallok 'res_area'-t / 'res_polys_gdf'-et / 'keep_polys'-t. "
        "Elbb futtasd azt a cellt, ami az sszes relevns residential poligont ellltja."
    )

def extract_lines(geom):
    if geom is None or geom.is_empty:
        return []
    gt = geom.geom_type
    if gt == "LineString":
        return [geom]
    if gt == "MultiLineString":
        return list(geom.geoms)
    if gt == "GeometryCollection":
        out = []
        for g in geom.geoms:
            out.extend(extract_lines(g))
        return out
    return []

def clip_lines(lines, poly):
    out = []
    for ln in lines:
        if ln is None or ln.is_empty:
            continue
        cut = ln.intersection(poly)
        out.extend(extract_lines(cut))
    return [g for g in out if g is not None and not g.is_empty]

# 1) vgand rtegek sszegyjtse (boundary NEM vgand)
street_lines = [g for g in edges.geometry if g is not None and not g.is_empty]

orange_lines = []
if "orange" in globals() and orange is not None and len(orange):
    orange_lines = [g for g in orange.geometry if g is not None and not g.is_empty]

blue_lines = []
if "blue" in globals() and blue is not None and len(blue):
    blue_lines = [g for g in blue.geometry if g is not None and not g.is_empty]

# 2) levgs az SSZES relevns lakott terletre: utck + narancs + kk
clipped_other = clip_lines(street_lines + orange_lines + blue_lines, clip_poly)
print("Levgott (utck+narancs+kk) vonalak:", len(clipped_other))

# 3) boundary hozzadsa (vltozatlanul)
boundary_lines = [g for g in boundary_gs.geometry if g is not None and not g.is_empty]
print("Boundary elemek:", len(boundary_lines))

# 4) VGS EGYRTEG HL: union + linemerge
all_final = clipped_other + boundary_lines
if not all_final:
    raise RuntimeError("Nincs semmi a vgs hlhoz (all_final res).")

try:
    u = unary_union(all_final)
    u_lines = extract_lines(u)
    merged_geom = linemerge(u_lines) if u_lines else u
    final_lines = extract_lines(merged_geom) or u_lines or all_final
except Exception as e:
    print("Union/merge hiba:", repr(e))
    final_lines = all_final

network_gs_proj = gpd.GeoSeries(final_lines, crs=edges.crs)
print("Vgs geom_type:\n", network_gs_proj.geom_type.value_counts())

# 5) Plot: egyetlen vonalrteg, egy szn
fig, ax = plt.subplots(figsize=(10, 10))
network_gs_proj.plot(ax=ax, linewidth=2.0, color="red")

ax.set_aspect("equal")
ax.set_facecolor("0.2")
ax.grid(True, color="0.4", linestyle="--", linewidth=0.5)

# zoom az SSZES lakott terlet unijra
minx, miny, maxx, maxy = clip_poly.bounds
side = max(maxx - minx, maxy - miny)
pad  = side * 0.05
ax.set_xlim(minx - pad, maxx + pad)
ax.set_ylim(miny - pad, maxy + pad)

plt.tight_layout()
plt.show()
```

```python
from shapely.ops import unary_union, linemerge, snap, nearest_points
from shapely.geometry import LineString, MultiLineString, GeometryCollection, Point
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

# -------------------------------------------------
# PARAMTEREK
# -------------------------------------------------
DO_SNAP_STRIP = True   # <<< ha nem kell a boundary menti "dupla fal" kezels, lltsd False-ra

# SNAP_TOL (mter)
# -----------------------------------------------
# A hlzat csompontjait s vonalvgeit ennyire "rhzza"
# a lakott terlet hatrvonalra (boundary).
#
# Mire j:
# - Ha az utca szinte rinti a boundary-t, de nem pontosan,
#   akkor ezzel tnylegesen rktjk.
#
# Tnet -> mit llts:
# - Ha az utck nem csatlakoznak a boundary-hez: NVELD (pl. 58 m)
# - Ha tl sok pont "elcsszik": CSKKENTSD (pl. 12 m)

SNAP_TOL = 3.0


# STRIP_TOL (mter)
# -----------------------------------------------
# A boundary krl ekkora szlessg svot kivgunk
# a bels thlzatbl, hogy eltntessk a prhuzamos
# "dupla fal" vonalakat.
#
# Mire j:
# - Ha a boundary mellett fut egy msik, szinte fedsben lv utca,
#   akkor azt eltnteti.
#
# Tnet -> mit llts:
# - Ha mg mindig van dupla vonal: NVELD (pl. 34 m)
# - Ha eltnnek fontos utck: CSKKENTSD (pl. 11.5 m)

STRIP_TOL = 10.0


# JOIN_TOL (mter)
# -----------------------------------------------
# Ha egy vonalvgpont ennyire kzel van a boundary-hez,
# akkor EXPLICIT sszekt szegmenst rajzolunk kzjk.
#
# Mire j:
# - Biztostja, hogy a hlzat tnylegesen kapcsoldjon
#   a boundary-hez, ne csak "majdnem".
#
# Tnet -> mit llts:
# - Ha maradnak rsek a poligon zrsnl: NVELD (pl. 812 m)
# - Ha felesleges, hossz rktsek jelennek meg: CSKKENTSD (pl. 45 m)

JOIN_TOL = 12.0


# DEDUP_EPS (mter)
# -----------------------------------------------
# Az egymshoz nagyon kzeli vgpontokat egynek vesszk,
# hogy ne rajzoljon tbb szinte azonos sszekt vonalat.
#
# Mire j:
# - Megakadlyozza a "tsks" hatrvonalat.
#
# Tnet -> mit llts:
# - Ha sok egyms melletti kis sszekts van: NVELD (0.51.0)
# - Ha fontos pontok elvesznek: CSKKENTSD (0.10.2)

DEDUP_EPS = 1.0


# -------------------------------------------------
# BEMENETEK (elz cellkbl)
# edges (GeoDataFrame, projektlt CRS)
# orange (GeoSeries) - opcionlis
# blue   (GeoSeries) - opcionlis
#
# s EGYIK ezek kzl:
#  - res_area (Polygon/MultiPolygon)  <-- ajnlott (sszes relevns lakott folt unija)
#  - res_polys_gdf / keep_polys (ebbl union-t csinlunk)
#  - boundary_gs + best_poly (rgi, de ez az eset NEM j tbb foltra)
# -------------------------------------------------

def extract_lines(geom):
    if geom is None or geom.is_empty:
        return []
    gt = geom.geom_type
    if gt == "LineString":
        return [geom]
    if gt == "MultiLineString":
        return list(geom.geoms)
    if gt == "GeometryCollection":
        out = []
        for g in geom.geoms:
            out.extend(extract_lines(g))
        return out
    return []

def clip_lines(lines, poly):
    out = []
    for ln in lines:
        if ln is None or ln.is_empty:
            continue
        cut = ln.intersection(poly)
        out.extend(extract_lines(cut))
    return [g for g in out if g is not None and not g.is_empty]

def endpoints_of_lines(lines):
    pts = []
    for ln in lines:
        if ln is None or ln.is_empty:
            continue
        c = list(ln.coords)
        if len(c) >= 2:
            pts.append(Point(c[0]))
            pts.append(Point(c[-1]))
    return pts

def dedup_points(points, eps):
    kept = []
    for p in points:
        ok = True
        for q in kept:
            if p.distance(q) <= eps:
                ok = False
                break
        if ok:
            kept.append(p)
    return kept

# -------------------------------------------------
# 0) CLIP POLY (MINDEN folt!)
# -------------------------------------------------
if "res_area" in globals() and res_area is not None:
    clip_poly = res_area
elif "res_polys_gdf" in globals() and res_polys_gdf is not None and len(res_polys_gdf):
    clip_poly = unary_union(res_polys_gdf.geometry)
elif "keep_polys" in globals() and keep_polys is not None and len(keep_polys):
    clip_poly = unary_union(keep_polys)
else:
    raise RuntimeError(
        "Nem tallok 'res_area'-t / 'res_polys_gdf'-et / 'keep_polys'-t. "
        "Elbb futtasd azt a cellt, ami az sszes relevns residential poligont ellltja."
    )

# boundary-t a clip_poly-bl szmoljuk (egysges, minden foltra)
boundary_line = clip_poly.boundary
boundary_lines = extract_lines(boundary_line)
if not boundary_lines:
    raise RuntimeError("Nem tudtam boundary vonalakat kinyerni (boundary_lines res).")

# -------------------------------------------------
# 1) sszegyjts: vgand rtegek (boundary-t NEM vgjuk)
# -------------------------------------------------
street_lines = [g for g in edges.geometry if g is not None and not g.is_empty]

orange_lines = []
if "orange" in globals() and orange is not None and len(orange):
    orange_lines = [g for g in orange.geometry if g is not None and not g.is_empty]

blue_lines = []
if "blue" in globals() and blue is not None and len(blue):
    blue_lines = [g for g in blue.geometry if g is not None and not g.is_empty]

# -------------------------------------------------
# 2) Levgs MINDEN lakott foltra: utck + narancs + kk
# -------------------------------------------------
clipped_other = clip_lines(street_lines + orange_lines + blue_lines, clip_poly)
print("Levgott (utck+narancs+kk) vonalak:", len(clipped_other))
print("Boundary elemek:", len(boundary_lines))

# -------------------------------------------------
# 3) (Opcionlis) boundary melletti dupla-fal kezels
# -------------------------------------------------
connectors = []

if DO_SNAP_STRIP and clipped_other:
    other_union = unary_union(clipped_other)
    if other_union and (not other_union.is_empty):

        # SNAP
        other_snapped = snap(other_union, boundary_line, SNAP_TOL)
        snapped_lines = extract_lines(other_snapped)

        # strip-ben fut rszek vgpontjai -> boundary-re rhz connectorok
        border_strip = boundary_line.buffer(STRIP_TOL)

        in_strip = []
        for ln in snapped_lines:
            in_strip.extend(extract_lines(ln.intersection(border_strip)))

        strip_endpoints = dedup_points(endpoints_of_lines(in_strip), DEDUP_EPS)

        for p in strip_endpoints:
            d = p.distance(boundary_line)
            if np.isfinite(d) and (1e-9 < d <= JOIN_TOL):
                _, q = nearest_points(p, boundary_line)
                if q is not None and (not q.is_empty):
                    seg = LineString([p, q])
                    if seg.length > 1e-6:
                        connectors.append(seg)

        # dupla fal eltntets: vgjuk ki a strip-et a snapped hlbl
        other_clean = other_snapped.difference(border_strip)
        clipped_other = [g for g in extract_lines(other_clean) if g is not None and not g.is_empty]

print("DO_SNAP_STRIP =", DO_SNAP_STRIP)
print("Connectorok szma:", len(connectors))
print("clipped_other (snap/strip utn):", len(clipped_other))

# -------------------------------------------------
# 4) Vgs EGY rteg: (clipped_other + connectors + boundary) -> union + linemerge
# -------------------------------------------------
all_final = clipped_other + connectors + boundary_lines
if not all_final:
    raise RuntimeError("Nincs semmi a vgs hlhoz (all_final res).")

try:
    u = unary_union(all_final)  # noding is itt trtnik
    u_lines = extract_lines(u)
    merged_geom = linemerge(u_lines) if u_lines else u
    final_lines = extract_lines(merged_geom) or u_lines or all_final
except Exception as e:
    print("Union/merge hiba:", repr(e))
    final_lines = all_final

network_gs_proj = gpd.GeoSeries(final_lines, crs=edges.crs)
print("Vgs geom_type:\n", network_gs_proj.geom_type.value_counts())

# -------------------------------------------------
# 5) Plot: egyetlen vonalrteg
# -------------------------------------------------
fig, ax = plt.subplots(figsize=(10, 10))
network_gs_proj.plot(ax=ax, linewidth=2.0, color="red")

ax.set_aspect("equal")
ax.set_facecolor("0.2")
ax.grid(True, color="0.4", linestyle="--", linewidth=0.5)

# zoom az SSZES lakott foltra
minx, miny, maxx, maxy = clip_poly.bounds
side = max(maxx - minx, maxy - miny)
pad  = side * 0.05
ax.set_xlim(minx - pad, maxx + pad)
ax.set_ylim(miny - pad, maxy + pad)

plt.tight_layout()
plt.show()
```

```python
from shapely.ops import unary_union, polygonize
from shapely.geometry import Polygon, MultiPolygon
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 0) BEMENET
# ------------------------------------------------------------
if "network_gs_proj" not in globals():
    raise RuntimeError("network_gs_proj nem ltezik  elbb futtasd a hlzatpt cellt.")

# ------------------------------------------------------------
# PARAMTEREK
# ------------------------------------------------------------
MIN_AREA = 5000.0      # m: ez alatt beolvasztjuk
MAX_STEPS = 20000    # biztonsgi limit (nagy hlnl se szlljon el)
PLOT_DEBUG = True

# ------------------------------------------------------------
# 1) VONALHL -> POLIGONOK
# ------------------------------------------------------------
linework = unary_union([g for g in network_gs_proj.geometry if g is not None and (not g.is_empty)])
polys = list(polygonize(linework))
if not polys:
    raise RuntimeError("polygonize nem adott vissza poligonokat (nincs elg zrt hurok / noding problma).")

polygons_gdf = gpd.GeoDataFrame(geometry=polys, crs=network_gs_proj.crs).reset_index(drop=True)

# tisztts (csak minimlisan, hogy ne "torztsunk")
# buffer(0) itt csak validlsra: nem hasznlunk tolerancis szomszdkeresst!
polygons_gdf["geometry"] = polygons_gdf.geometry.buffer(0)
polygons_gdf = polygons_gdf[polygons_gdf.geometry.type.isin(["Polygon", "MultiPolygon"])].reset_index(drop=True)

print("Polygonok szma (polygonize utn):", len(polygons_gdf))
print("Min area (m):", float(polygons_gdf.geometry.area.min()))
print("< MIN_AREA db:", int((polygons_gdf.geometry.area < MIN_AREA).sum()))

# eredeti lefedettsg (ellenrzshez)
orig_union = unary_union(polygons_gdf.geometry)

# ------------------------------------------------------------
# 2) KICS I POLIGONOK BEOLVASZTSA (EGYENKNT)
#    - csak VALDI szomszd: kzs hatrhossz > 0
#    - cl: akivel a leghosszabb a kzs hatr
# ------------------------------------------------------------
pg = polygons_gdf.copy().reset_index(drop=True)

def shared_boundary_length(a, b):
    inter = a.boundary.intersection(b.boundary)
    return getattr(inter, "length", 0.0)

steps = 0
while steps < MAX_STEPS:
    steps += 1

    areas = pg.geometry.area
    small_idx = areas[areas < MIN_AREA].index.tolist()
    if not small_idx:
        break

    # mindig a legkisebbet olvasztjuk be elszr (stabilabb)
    i = min(small_idx, key=lambda k: areas.iloc[k])
    gi = pg.geometry.iloc[i]

    # szomszdkeress bbox + valdi kzs hatr
    sidx = pg.sindex
    cand = list(sidx.intersection(gi.bounds))
    cand = [j for j in cand if j != i]

    best_j = None
    best_len = 0.0

    for j in cand:
        gj = pg.geometry.iloc[j]
        L = shared_boundary_length(gi, gj)
        if L > best_len:
            best_len = L
            best_j = j

    # Ha nincs valdi szomszd kzs llel, akkor nem tudjuk szablyosan beolvasztani
    # (ez tipikusan azt jelenti, hogy a polygonize partciban van mikro rs / diszkontinuits)
    if best_j is None or best_len <= 0:
        print(f"[STOP] Kicsi poligon ({i}, area={areas.iloc[i]:.6f}) nem tall valdi szomszdot kzs llel.")
        break

    # olvaszts: i -> best_j
    new_geom = unary_union([gi, pg.geometry.iloc[best_j]]).buffer(0)

    # frissts: clpoligon helyre j geom, kicsit eldobjuk
    pg.at[best_j, "geometry"] = new_geom
    pg = pg.drop(index=i).reset_index(drop=True)

print("Merge lpsek:", steps)
print("Vgs poligonok:", len(pg))
print("Vgs min area (m):", float(pg.geometry.area.min()))
print("Vgs < MIN_AREA db:", int((pg.geometry.area < MIN_AREA).sum()))

polygons_merged_gdf = pg  # <- ezt hasznld tovbb

# ------------------------------------------------------------
# 3) KEMNY ELLENRZSEK: nincs tfeds, nincs terletveszts
# ------------------------------------------------------------
final_union = unary_union(polygons_merged_gdf.geometry)

sum_area = float(polygons_merged_gdf.geometry.area.sum())
union_area = float(final_union.area)
overlap_area = sum_area - union_area   # ha > 0, akkor tfedsek vannak

symdiff_area = float(orig_union.symmetric_difference(final_union).area)  # ha > 0, akkor veszts/hozzads trtnt

print("Ellenrzs:")
print("  sum(area) - union(area) (tfeds indiktor):", overlap_area)
print("  symmetric_difference area (terlet eltrs):", symdiff_area)

# ------------------------------------------------------------
# 4) PLOT
# ------------------------------------------------------------
if PLOT_DEBUG:
    fig, ax = plt.subplots(figsize=(10, 10))
    polygons_merged_gdf.plot(ax=ax, facecolor="none", edgecolor="cyan", linewidth=1.0)
    ax.set_aspect("equal")
    ax.set_facecolor("black")
    ax.set_axis_off()
    plt.tight_layout()
    plt.show()

print("Ksz: polygons_merged_gdf  kicsik egyenknt beolvasztva a leghosszabb kzs l szomszdba.")
```
