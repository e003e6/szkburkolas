```python
# 2. szabály

_ADDR = re.compile(r"^\s*(\d+)\s*(?:-\s*(\d+))?\s*([A-Za-z])?\s*$")

def parse_addr(s):
    '''
    szétszedi a cím bemenetet három tagra 12-14 C -> (12, 14, 'C')
    ha nem tudja megcsinálni akkor None
    '''
    
    if s is None or (isinstance(s, float) and pd.isna(s)):
        return None
        
    m = _ADDR.match(str(s))
    
    if not m:
        return None
        
    a = int(m.group(1))
    b = int(m.group(2)) if m.group(2) else a
    lo, hi = (a, b) if a <= b else (b, a)
    letter = m.group(3).upper() if m.group(3) else None
    
    return (lo, hi, letter)


def overlap(a, b):
    if a[2] is not None or b[2] is not None:
        if a[2] != b[2]:
            return False
    return max(a[0], b[0]) <= min(a[1], b[1])


def attach_by_range(df_unmatched, gdf, df_id_col="_df_rowid"):
    d = df_unmatched.copy()
    g = gdf.copy()

    if df_id_col not in d.columns:
        d[df_id_col] = range(len(d))

    # df-ből dobd a korábbról beragadt gdf-es oszlopokat
    d = d.drop(columns=[c for c in d.columns if "_gdf" in c], errors="ignore")
    # df oldali üres geometry ne zavarjon be
    d = d.drop(columns=["geometry"], errors="ignore")

    # cím parse: kizárjuk a visszafelé tartományt (b <= a)
    def _parse(s):
        if s is None or (isinstance(s, float) and pd.isna(s)):
            return None
        m = _ADDR.match(str(s))
        if not m:
            return None
        a = int(m.group(1))
        b_raw = m.group(2)
        if b_raw is not None:
            b = int(b_raw)
            if b <= a:         # <-- ez az új szabály
                return None
        else:
            b = a
        L = m.group(3).upper() if m.group(3) else None
        return (a, b, L)

    d["_a"] = d["cim"].map(_parse)
    g["_b"] = g["cim"].map(_parse)

    d = d[d["_a"].notna()]
    g = g[g["_b"].notna()]

    if d.empty or g.empty:
        empty = gpd.GeoDataFrame(d.iloc[0:0].copy(), geometry=None, crs=gdf.crs)
        still = d.drop(columns=["_a", df_id_col], errors="ignore")
        return empty, still

    d[["d_lo", "d_hi", "d_L"]] = pd.DataFrame(d["_a"].tolist(), index=d.index)
    g[["g_lo", "g_hi", "g_L"]] = pd.DataFrame(g["_b"].tolist(), index=g.index)

    # gdf kulcsok átnevezése, hogy megmaradjon minden
    g = g.rename(columns={"utca": "utca_gdf", "cim": "cim_gdf"})

    cand = d.merge(
        g,
        how="left",
        left_on=["utca", "telepulesnev"],
        right_on=["utca_gdf", "telepules"],
        suffixes=("_df", "_gdf"),
    )

    # numerikus átfedés
    num_ok = cand[["d_lo", "g_lo"]].max(axis=1) <= cand[["d_hi", "g_hi"]].min(axis=1)

    # betű szabály: ha bármelyiken van betű, egyezzen
    letter_needed = cand["d_L"].notna() | cand["g_L"].notna()
    letter_ok = ~letter_needed | (cand["d_L"] == cand["g_L"])

    m = num_ok & letter_ok & cand["g_lo"].notna()

    matched_df = cand[m].drop(
        columns=["_a", "_b", "d_lo", "d_hi", "d_L", "g_lo", "g_hi", "g_L"],
        errors="ignore",
    )

    # geometry oszlop kiválasztása (gdf-ből jövő)
    if "geometry_gdf" in matched_df.columns:
        geom_col = "geometry_gdf"
    elif "geometry" in matched_df.columns:
        geom_col = "geometry"
    else:
        geom_col = next((c for c in matched_df.columns if c.startswith("geometry") and "_gdf" in c), None)

    matched = gpd.GeoDataFrame(matched_df, geometry=geom_col, crs=gdf.crs) if geom_col else gpd.GeoDataFrame(matched_df, geometry=None, crs=gdf.crs)

    matched_ids = matched[df_id_col].unique()
    still_unmatched = d[~d[df_id_col].isin(matched_ids)].drop(
        columns=["_a", "d_lo", "d_hi", "d_L"],
        errors="ignore",
    )

    # kimenet takarítás: technikai oszlopok ki
    cleanup = [df_id_col]
    matched = matched.drop(columns=cleanup, errors="ignore")
    still_unmatched = still_unmatched.drop(columns=cleanup, errors="ignore")

    return matched, still_unmatched
```

```python
def attach_ignore_letter(df_unmatched, gdf, df_id_col="_df_rowid"):
    d = df_unmatched.copy()
    g = gdf.copy()

    if df_id_col not in d.columns:
        d[df_id_col] = range(len(d))

    # ha beragadt gdf-es oszlopok vannak a df_unmatched-ben, dobd ki
    d = d.drop(columns=[c for c in d.columns if "_gdf" in c], errors="ignore")
    d = d.drop(columns=["geometry"], errors="ignore")

    # csak szám/tartomány kell, betűt ignoráljuk
    # regex: 1) lo  2) hi(optional)  3) letter(optional)
    ext_d = d["cim"].astype("string").str.extract(_ADDR)
    ext_g = g["cim"].astype("string").str.extract(_ADDR)

    # lo/hi számokká
    d_lo = pd.to_numeric(ext_d[0], errors="coerce")
    d_hi = pd.to_numeric(ext_d[1], errors="coerce")
    g_lo = pd.to_numeric(ext_g[0], errors="coerce")
    g_hi = pd.to_numeric(ext_g[1], errors="coerce")

    # ha nincs hi, akkor hi=lo (pont)
    d_hi = d_hi.fillna(d_lo)
    g_hi = g_hi.fillna(g_lo)

    # kizárjuk a visszafelé tartományt: ahol volt hi és hi <= lo
    d_has_hi = ext_d[1].notna()
    g_has_hi = ext_g[1].notna()
    d_ok = d_lo.notna() & d_hi.notna() & (~d_has_hi | (d_hi > d_lo))
    g_ok = g_lo.notna() & g_hi.notna() & (~g_has_hi | (g_hi > g_lo))

    d = d[d_ok].copy()
    g = g[g_ok].copy()

    if d.empty or g.empty:
        empty = gpd.GeoDataFrame(d.iloc[0:0].copy(), geometry=None, crs=gdf.crs)
        still = d.drop(columns=[df_id_col], errors="ignore")
        return empty, still

    # a szűrt indexek alapján töltsük vissza a lo/hi-t
    d["d_lo"] = d_lo.loc[d.index].astype(int)
    d["d_hi"] = d_hi.loc[d.index].astype(int)
    g["g_lo"] = g_lo.loc[g.index].astype(int)
    g["g_hi"] = g_hi.loc[g.index].astype(int)

    # gdf kulcsok átnevezése, hogy minden megmaradjon
    g = g.rename(columns={"utca": "utca_gdf", "cim": "cim_gdf"})

    cand = d.merge(
        g,
        how="left",
        left_on=["utca", "telepulesnev"],
        right_on=["utca_gdf", "telepules"],
        suffixes=("_df", "_gdf"),
    )

    # numerikus átfedés
    num_ok = cand[["d_lo", "g_lo"]].max(axis=1) <= cand[["d_hi", "g_hi"]].min(axis=1)
    m = num_ok & cand["g_lo"].notna()

    matched_df = cand[m].drop(columns=["d_lo", "d_hi", "g_lo", "g_hi"], errors="ignore")

    # geometry oszlop (gdf-ből jövő)
    if "geometry_gdf" in matched_df.columns:
        geom_col = "geometry_gdf"
    elif "geometry" in matched_df.columns:
        geom_col = "geometry"
    else:
        geom_col = next((c for c in matched_df.columns if c.startswith("geometry") and "_gdf" in c), None)

    matched = gpd.GeoDataFrame(matched_df, geometry=geom_col, crs=gdf.crs) if geom_col else gpd.GeoDataFrame(matched_df, geometry=None, crs=gdf.crs)

    matched_ids = matched[df_id_col].unique()
    still_unmatched = d[~d[df_id_col].isin(matched_ids)].drop(columns=["d_lo", "d_hi"], errors="ignore")

    # technikai oszlop kidobása a kimenetből
    matched = matched.drop(columns=[df_id_col], errors="ignore")
    still_unmatched = still_unmatched.drop(columns=[df_id_col], errors="ignore")

    return matched, still_unmatched
```
