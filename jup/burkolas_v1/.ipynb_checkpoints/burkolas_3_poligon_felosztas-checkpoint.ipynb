```python
# pip install geopandas shapely scipy pandas numpy matplotlib

import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, Point, LineString
from scipy.optimize import linprog

# -----------------------------
# 0) Poligon (településhatár jellegű)
# -----------------------------
poly = Polygon([
    (0, 0), (10, 0), (12, 3), (11, 7),
    (8, 10), (3, 11), (0, 8), (-1, 3)
])

# ---------------------------------------------------
# 1) "Ház koordináták" generálása (rács + kis zaj)
# ---------------------------------------------------
rng = np.random.default_rng(42)

def grid_points_in_polygon(
    poly: Polygon,
    x_range: tuple[float, float],
    y_range: tuple[float, float],
    step: float,
    jitter: float
):
    pts = []
    xs = np.arange(x_range[0], x_range[1], step)
    ys = np.arange(y_range[0], y_range[1], step)

    for x in xs:
        for y in ys:
            jx = x + rng.uniform(-jitter, jitter)
            jy = y + rng.uniform(-jitter, jitter)
            p = Point(jx, jy)
            if poly.contains(p):
                pts.append(p)
    return pts

# A: "felső városrész"
A = grid_points_in_polygon(
    poly,
    x_range=(1.5, 5.5),
    y_range=(6.0, 10.0),
    step=0.6,
    jitter=0.08
)

# B: "alsó városrész"
B = grid_points_in_polygon(
    poly,
    x_range=(6.5, 10.5),
    y_range=(1.5, 5.0),
    step=0.6,
    jitter=0.08
)

# pár hibás pont (rossz címkézés / zaj)
A.append(B[3])
B.append(A[5])

gdfA = gpd.GeoDataFrame({"cls": ["A"] * len(A)}, geometry=A, crs="EPSG:3857")
gdfB = gpd.GeoDataFrame({"cls": ["B"] * len(B)}, geometry=B, crs="EPSG:3857")
gdf = gpd.GeoDataFrame(pd.concat([gdfA, gdfB], ignore_index=True), crs=gdfA.crs)

# ---------------------------------------------------
# 2) Soft-margin max-margin szeparátor (hibatűrő)
# ---------------------------------------------------
X = np.array([[p.x, p.y] for p in gdf.geometry])
y = np.where(gdf["cls"].to_numpy() == "A", 1.0, -1.0)

n = len(X)
c = np.zeros(4 + n)
C = 1.0   # hibatűrés mértéke (városi adatoknál ez jó indulás)
c[3] = 1.0
c[4:] = C

A_ub = []
b_ub = []

for i in range(n):
    row = np.zeros(4 + n)
    row[0] = -y[i] * X[i, 0]
    row[1] = -y[i] * X[i, 1]
    row[2] = -y[i]
    row[4 + i] = -1.0
    A_ub.append(row)
    b_ub.append(-1.0)

for sign, idx in [(1,0), (-1,0), (1,1), (-1,1)]:
    row = np.zeros(4 + n)
    row[idx] = sign
    row[3] = -1
    A_ub.append(row)
    b_ub.append(0)

res = linprog(
    c,
    A_ub=np.vstack(A_ub),
    b_ub=np.array(b_ub),
    bounds=[(None,None),(None,None),(None,None),(0,None)] + [(0,None)]*n,
    method="highs"
)

w_x, w_y, b0 = res.x[:3]

# -----------------------------
# 3) Elválasztó vonal
# -----------------------------
minx, miny, maxx, maxy = poly.bounds
pad = 10
minx -= pad; maxx += pad; miny -= pad; maxy += pad

if abs(w_y) > 1e-9:
    p1 = (minx, -(w_x*minx + b0)/w_y)
    p2 = (maxx, -(w_x*maxx + b0)/w_y)
else:
    x0 = -b0 / w_x
    p1 = (x0, miny)
    p2 = (x0, maxy)

cut_line = LineString([p1, p2])

# -----------------------------
# 4) Plot Jupyterbe
# -----------------------------
fig, ax = plt.subplots(figsize=(8, 8))

# poligon
gpd.GeoSeries([poly]).plot(
    ax=ax,
    facecolor="none",
    edgecolor="black",
    linewidth=2
)

# pontok (házak)
gdfA.plot(ax=ax, markersize=25, label="A városrész", alpha=0.9)
gdfB.plot(ax=ax, markersize=25, label="B városrész", alpha=0.9)

# elválasztó vonal
gpd.GeoSeries([cut_line]).plot(
    ax=ax,
    linestyle="--",
    linewidth=2,
    label="elválasztó vonal"
)

ax.set_aspect("equal")
ax.legend()
ax.set_title("Ház jellegű koordináták – hibatűrő elválasztás")
plt.show()
```
