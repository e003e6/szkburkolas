## Utca hálózat, lakott terület határoló vonal és utcavégek összekötve határoló vonallal

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from shapely.ops import nearest_points
from shapely.geometry import LineString, Polygon, MultiPolygon

# -----------------------
# PARAMÉTEREK
# -----------------------

PLACE = "Tököl, Hungary"   # Vizsgált település neve az OSM-ben
BUFFER_M = 80              # Buffer-szélesség (méter) az utakon kívül
DIST_LIM = 100              # Max. távolság (méter) a csomópont és a határ között

# ÚJ: simítás mértéke méterben.
# Ez tölti be a bevágásokat / öblöket – a városhatár „görbületének” léptéke.
SMOOTH_M = 200             # Próbáld 200–400 között finomhangolni

# ÚJ: végső egyszerűsítés a recés vonal ellen (méterben).
SIMPLIFY_TOL_M = 50        # 30–80 m között érdemes próbálkozni


# -----------------------
# 1) ÚTHÁLÓZAT LETÖLTÉSE ÉS PROJEKTÁLÁSA
# -----------------------

G = ox.graph_from_place(PLACE, network_type="drive")
Gp = ox.project_graph(G)
nodes_gdf, edges_gdf = ox.graph_to_gdfs(Gp, nodes=True, edges=True)


# -----------------------
# 2) HATÁROLÓ VONAL KÉSZÍTÉSE – GYAKORLATI, SIMÍTOTT VÁROSHATÁR
# -----------------------

# 2.1. Nyers út-buffer (szögletes)
raw_buf = edges_gdf.geometry.buffer(
    BUFFER_M,
    cap_style=2,   # square végek
    join_style=2   # éles sarkok
).union_all()

# 2.2. Morfológiai simítás:
# Először kifelé buffereljük SMOOTH_M-mel (kitölti a keskeny bevágásokat),
# majd visszafelé ugyanennyivel (összehúzza, de az öblöket már eltüntette).
smoothed_buf = raw_buf.buffer(SMOOTH_M).buffer(-SMOOTH_M)

# 2.3. Opcionális finom simplifikálás, hogy ne legyen méterenként törés,
# de a város alakját még kövesse.
smoothed_buf = smoothed_buf.simplify(SIMPLIFY_TOL_M, preserve_topology=True)

# 2.4. Csak a külső poligon(ok) kezelése
polygons = []
if isinstance(smoothed_buf, Polygon):
    polygons = [smoothed_buf]
elif isinstance(smoothed_buf, MultiPolygon):
    polygons = list(smoothed_buf.geoms)

# Ha több poligon lenne (pl. mellékfalvak), általában a legnagyobb területű
# az, ami a fő „városhatárnak” tekinthető.
if polygons:
    main_poly = max(polygons, key=lambda p: p.area)
    exteriors = [main_poly.exterior]
else:
    exteriors = []

# A határvonal(ak) GeoSeries-ben, projektált CRS-ben
boundary_lines_proj = gpd.GeoSeries(exteriors, crs=edges_gdf.crs)

# A távolságszámításhoz kell egy összefogott vonalgeometria
boundary_mls_proj = boundary_lines_proj.unary_union


# -----------------------
# 3) CSOMÓPONTOK ÖSSZEKÖTÉSE A HATÁRVONALLAL
# -----------------------

extensions_proj = []

for idx, row in nodes_gdf.iterrows():
    pt = row.geometry

    d = pt.distance(boundary_mls_proj)
    if pd.notna(d) and np.isfinite(d) and d <= DIST_LIM:
        _, near = nearest_points(pt, boundary_mls_proj)
        if near and not near.is_empty:
            seg = LineString([pt, near])
            if seg.length > 0.1:
                extensions_proj.append(seg)


# -----------------------
# 4) VISSZAVETÍTÉS WGS84-BE
# -----------------------

boundary_lines_wgs = boundary_lines_proj.to_crs(epsg=4326)

if extensions_proj:
    extensions_wgs = gpd.GeoSeries(extensions_proj, crs=edges_gdf.crs).to_crs(epsg=4326)
else:
    extensions_wgs = gpd.GeoSeries([], crs=edges_gdf.crs).to_crs(epsg=4326)


# -----------------------
# 5) ÉL-SZÍNEZÉS NÉV ALAPJÁN
# -----------------------

def has_name(val):
    if isinstance(val, list):
        return any(isinstance(i, str) and i.strip() for i in val)
    if pd.isna(val):
        return False
    return isinstance(val, str) and val.strip()

edge_colors = [
    "white" if has_name(data.get("name")) else "red"
    for _, _, _, data in G.edges(keys=True, data=True)
]


# -----------------------
# 6) RAJZOLÁS
# -----------------------

fig, ax = ox.plot_graph(
    G,
    node_size=0,
    edge_color=edge_colors,
    bgcolor="black",
    show=False,
    close=False
)

ax.set_aspect("auto")

if len(boundary_lines_wgs):
    boundary_lines_wgs.plot(ax=ax, color="lime", linewidth=2)

if len(extensions_wgs):
    extensions_wgs.plot(ax=ax, color="dodgerblue", linewidth=1.8)

if len(boundary_lines_wgs):
    minx, miny, maxx, maxy = boundary_lines_wgs.total_bounds
    pad = 0.002
    ax.set_xlim(minx - pad, maxx + pad)
    ax.set_ylim(miny - pad, maxy + pad)

plt.show()
```

## A három vonalak metszetéből vonalak egyesítése egyetelen vonalréteggé

```python
from shapely.ops import unary_union, linemerge
from shapely.geometry import LineString, MultiLineString, GeometryCollection
import geopandas as gpd
import matplotlib.pyplot as plt

# Feltételezzük, hogy ezek már léteznek a notebookban:
# edges_gdf (projektált CRS-ben)
# boundary_lines_proj (projektált CRS-ben)
# extensions_proj (lista LineString-ekkel, projektált CRS-ben)

# 1) Vonalak összegyűjtése
street_lines    = [g for g in edges_gdf.geometry if g is not None and not g.is_empty]
boundary_lines  = [g for g in boundary_lines_proj.geometry if g is not None and not g.is_empty]
connector_lines = [g for g in extensions_proj if g is not None and not g.is_empty]

all_lines_proj = street_lines + boundary_lines + connector_lines
print("Összes vonal:", len(all_lines_proj))

if not all_lines_proj:
    raise RuntimeError("Nincs egyesíthető vonal (all_lines_proj üres).")

def extract_lines(geom):
    """Kiszed minden LineString-et / MultiLineString-et egy tetszőleges geometriából."""
    lines = []
    if geom is None or geom.is_empty:
        return lines

    if isinstance(geom, LineString):
        lines.append(geom)
    elif isinstance(geom, MultiLineString):
        lines.extend(list(geom.geoms))
    elif isinstance(geom, GeometryCollection):
        for g in geom.geoms:
            lines.extend(extract_lines(g))
    return lines

# 2) Topológiai egyesítés (ha sikerül)
try:
    union_geom  = unary_union(all_lines_proj)
    union_lines = extract_lines(union_geom)
    print("unary_union után vonalak:", len(union_lines))

    if union_lines:
        merged_geom  = linemerge(union_lines)
        merged_lines = extract_lines(merged_geom)
        print("linemerge után vonalak:", len(merged_lines))
    else:
        merged_lines = []
except Exception as e:
    print("Hiba a unary_union / linemerge során:", repr(e))
    union_lines  = []
    merged_lines = []

# 3) Végső hálózat kiválasztása
if merged_lines:
    final_lines = merged_lines
    print("Végső hálózat: merged_lines")
elif union_lines:
    final_lines = union_lines
    print("Végső hálózat: union_lines")
else:
    final_lines = all_lines_proj
    print("Végső hálózat: all_lines_proj (nem sikerült az egyesítés)")

network_gs_proj = gpd.GeoSeries(final_lines, crs=edges_gdf.crs)
print("geom_type:\n", network_gs_proj.geom_type.value_counts())
print("total_bounds:", network_gs_proj.total_bounds)


# 4) Plot – CSAK az egyesített / kiválasztott vonalhálózat

fig, ax = plt.subplots(figsize=(10, 10))

# Nagyon látványos stílus: VASTAG PIROS VONALAK, SZÜRKE HÁTTÉR
network_gs_proj.plot(ax=ax, linewidth=2.0, color="red")

ax.set_aspect("equal")      # valódi arányok
ax.set_facecolor("0.2")     # sötétszürke háttér, hogy a piros nagyon látszódjon
ax.grid(True, color="0.4", linestyle="--", linewidth=0.5)  # rács, hogy biztos lásd a teret

# Rázúm a hálózatra
minx, miny, maxx, maxy = network_gs_proj.total_bounds
side = max(maxx - minx, maxy - miny)
pad  = side * 0.05
ax.set_xlim(minx - pad, maxx + pad)
ax.set_ylim(miny - pad, maxy + pad)

plt.tight_layout()
plt.show()
```

## A vonalakból poligonok készítése, minden metszet egy külön poligon

```python
from shapely.ops import unary_union, polygonize
from shapely.geometry import LineString, MultiLineString, GeometryCollection
import geopandas as gpd
import matplotlib.pyplot as plt

# ------------------------------------------------------------------
# 0) BEMENET: network_gs_proj
#    (egyesített, érintkező vonalhálózat PROJEKTÁLT CRS-ben)
# ------------------------------------------------------------------

if "network_gs_proj" not in globals():
    raise RuntimeError("network_gs_proj nem létezik – előbb futtasd a hálózatépítő cellát.")

print("Vonalak száma a hálózatban:", len(network_gs_proj))

# Csak érvényes, nem üres vonalak
lines = [g for g in network_gs_proj.geometry if isinstance(g, LineString) and not g.is_empty]

if not lines:
    raise RuntimeError("Nincsenek érvényes LineString geometriák a network_gs_proj-ban.")


# ------------------------------------------------------------------
# 1) VONALHÁLÓZAT EGYESÍTÉSE (BIZTOS TOPOLÓGIA ELŐTT)
# ------------------------------------------------------------------

# unary_union: egyetlen topológiailag konzisztens lineworkké alakítja,
# közös csomópontokkal, széttördelt szegmensekkel
union_geom = unary_union(lines)

# Biztonsági diagnosztika
print("union_geom típusa:", type(union_geom))


# ------------------------------------------------------------------
# 2) POLIGONIZÁLÁS:
#    Minden zárt "cellát" / szigetet poligonná alakítunk
# ------------------------------------------------------------------

# polygonize: a vonalhálózat zárt gyűrűiből poligonokat hoz létre
polygons = list(polygonize(union_geom))

print("Keletkezett poligonok száma:", len(polygons))

# GeoDataFrame a poligonokról (külső gyűrűk adta cellák)
polygons_gdf = gpd.GeoDataFrame(geometry=polygons, crs=network_gs_proj.crs)

# Ha szeretnél, itt opcionálisan kiszűrheted a nagyon apró "zajos" poligonokat:
# min_area = 1000.0  # pl. 1000 m²
# polygons_gdf = polygons_gdf[polygons_gdf.area > min_area]


# ------------------------------------------------------------------
# 3) EREDMÉNY ELLENŐRZÉSE: PLOT
# ------------------------------------------------------------------

fig, ax = plt.subplots(figsize=(10, 10))

# Poligonok: csak kontúr (átlátszó kitöltés)
polygons_gdf.plot(ax=ax, facecolor="none", edgecolor="blue", linewidth=1.0)

# Hálózat: háttérként vékony fehér vonalak
network_gs_proj.plot(ax=ax, color="white", linewidth=0.3)

ax.set_aspect("equal")
ax.set_facecolor("black")
ax.set_axis_off()

plt.tight_layout()
plt.show()

# Plusz infó, hogy tudd, mivel tudsz tovább dolgozni
print("polygons_gdf készen van, egy sor = egy zárt poligon (belső sziget a vonalhálóban).")
```
